```python
x_list = df['Input'].tolist()

print(x_list)
```

```python
change=[]
for i in range(1,len(x_list)):
  diff=x_list[i]-x_list[i-1]
  rounded_val=round(diff,2)
  change.append(rounded_val)
print(change)
  
```

```python
gain=[]
for i in change:
  result = i if i > 0 else 0
  gain.append(result)
print(gain)
```

```python
loss=[]
for i in change:
  result = -i if i < 0 else 0
  loss.append(result)
print(loss)
```

```python
avg_gain=[]
new_vals=gain[0:14]
result = sum(gain[0:14])/14

rounded_val1=round(result,2)
avg_gain.append(rounded_val1)
#avg_gain
gain_list1=gain[14:]
gain_list1

for i in range(0,18):
  results=avg_gain
  result = ((avg_gain[i] * 13) + gain_list1[i]) / 14
  #print(result)
  rounded_val2=round(result,2)
  avg_gain.append(rounded_val2)
  #print(result)
print(avg_gain)
```

```python
avg_loss=[]
new_vals=loss[0:14]
result = sum(loss[0:14])/14
#print(result)
rounded_val1=round(result,2)
avg_loss.append(rounded_val1)

loss_list1=loss[14:]#f17
loss_list1

for i in range(0,18):
  results=avg_loss
  result = ((avg_loss[i] * 13) + loss_list1[i]) / 14
  rounded_val2=round(result,2)
  avg_loss.append(rounded_val2)
  #print(result)
print(avg_loss)
  
```

```python
hm=[]
count=0
for i in avg_gain:
  result=i/avg_loss[count]
  count=count+1
  rounded_val3=round(result,2)
  hm.append(rounded_val3)
print(hm)
```

```python
values_h16 = avg_loss[0:19]
values_i16 = hm[0:19]
#print(values_i16)
results = []
for i in range(len(values_h16)):
    if values_h16[i] == 0:
        result = 100
    else:
        result = 100 - (100 / (1 + values_i16[i]))
    results.append(result)

print(results)
new_values = ["{:.2f}".format(value) for value in results]

print(new_values)
```
